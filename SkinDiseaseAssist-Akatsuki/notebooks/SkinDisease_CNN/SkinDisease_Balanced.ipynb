```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
```

```python
data_dir = "./IMG_CLASSES"
import pathlib
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
classes = ["ez", "ml", "ad", "bcc", "mn", "bkl", "lp", "sk", "trc", "wm"]

disease_type_dict = {
    'ez': list(data_dir.glob('1. Eczema 1677/*')),
    'ml': list(data_dir.glob('2. Melanoma 15.75k/*')),
    'ad': list(data_dir.glob('3. Atopic Dermatitis - 1.25k/*')),
    'bcc': list(data_dir.glob('4. Basal Cell Carcinoma (BCC) 3323/*')),
    'mn': list(data_dir.glob('5. Melanocytic Nevi (NV) - 7970/*')),
    'bkl': list(data_dir.glob('6. Benign Keratosis-like Lesions (BKL) 2624/*')),
    'lp': list(data_dir.glob('7. Psoriasis pictures Lichen Planus and related diseases - 2k/*')),
    'sk': list(data_dir.glob('8. Seborrheic Keratoses and other Benign Tumors - 1.8k/*')),
    'trc': list(data_dir.glob('9. Tinea Ringworm Candidiasis and other Fungal Infections - 1.7k/*')),
    'wm': list(data_dir.glob('10. Warts Molluscum and other Viral Infections - 2103/*')),

}

labels = {
    'ez': 0,
    'ml': 1,
    'ad': 2,
    'bcc': 3,
    'mn': 4,
    'bkl': 5,
    'lp': 6,
    'sk': 7,
    'trc': 8,
    'wm': 9
}

[len(disease_type_dict[key]) for key in disease_type_dict.keys()]
```

```python
import random
limiter = 1200
for key in disease_type_dict.keys():
    newList = random.sample(disease_type_dict[key], limiter)
    disease_type_dict[key] = newList
    del newList

[len(disease_type_dict[key]) for key in disease_type_dict.keys()]
```

```python
X, y = [], []
for disease_name, disease_list in disease_type_dict.items():
    for diseaseImg in disease_list:
        print(len(X))
        #Fit image
        img = cv2.imread(str(diseaseImg))
        imgResized = cv2.resize(img,(180,180))

        X.append(imgResized)
        y.append(labels[disease_name])

X = np.array(X)
y = np.array(y)
```

```python
#CNN Hyperparameters
nEpochs = 8
nBatch = 32
nFolds = 5
```

```python
#Building CNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Dropout
from tensorflow.keras.layers import MaxPooling2D, Flatten

model = Sequential([
    #Preprocessor
    #dataAugment_preprocessor,

      Conv2D(64, (3, 3), activation="relu", padding="same", input_shape=(180, 180, 3)),
      MaxPooling2D(pool_size=(2, 2)),
      Dropout(0.20),

      Conv2D(64, (3, 3), activation="relu", padding="same"),
      MaxPooling2D(pool_size=(2, 2)),
      Dropout(0.20),

      Conv2D(32, (3, 3), activation="relu", padding="same"),
      MaxPooling2D(pool_size=(2, 2)),
      Dropout(0.20),

      Conv2D(32, (3, 3), activation="relu", padding="same"),
      MaxPooling2D(pool_size=(2, 2)),
      Dropout(0.20),

      Flatten(),

      Dense(units=256, activation="relu"),
      Dense(units=10, activation="softmax")
  ])
```

```python
#Fitting Data to Model
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.model_selection import StratifiedKFold
import time

hists_CNN = {}
fold = 1

skf = StratifiedKFold(n_splits=nFolds, random_state=7, shuffle=True).split(X, y)
start = time.time()

for train, test in skf:
  model.compile(loss=SparseCategoricalCrossentropy(), optimizer="adam", metrics=['accuracy'])
  model.summary()

  model_name = f"Models/model-kfold-epcohs-8-{fold}.h5"
  hist = model.fit(X[train], y[train],
                   epochs=nEpochs,
                   validation_split=0.2,
                   batch_size=nBatch,
                   verbose=1)

  model.save(model_name)
  pred = model.predict(X[test]).ravel()
  hists_CNN[f"history-fold-{fold}"] = hist
  fold+=1

end=time.time()
print(f"Training Time: {end - start}seconds")
```
